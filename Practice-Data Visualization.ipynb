```python
df = pd.read_csv("Car Accidents.csv")
bar_plot = df[df["Wind_Speed(mph)"]>30].plot.bar(x = "County", y= "Temperature(F)")
plt.show(bar_plot)
```

My bar plot uses my "Car Accidents.csv" which shows the many details of car acciddnts in counties all over the country. This chart shows the accidents where the temperature during the time of the accident went over 30 Fahrenheit. My conclusion from this data is that more accidents happen as the temperature lowers which can affect roads, visbility, etc.

```python
number_2 = sns.lmplot(x = "Severity" , y= "Temperature(F)",data= df,hue = "Precipitation(in)")
plt.show(number_2)
```
