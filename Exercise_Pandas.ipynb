```python
import pandas as pd
df1 = pd.read_csv("mens1500_2006.csv")
df2 = pd.read_csv("mens1500_2010.csv")
df3 = pd.read_csv("mens1500_2014.csv")
df4 = pd.read_csv("mens1500_2018.csv")
df5 = pd.read_csv("mens1500_2022.csv")

df = pd.concat([df1,df2,df3,df4,df5])
df
```

```python
filter = df.loc[:,"Athlete" : "Country"]
filter
```

I used the df.loc syntax to boolean index two columns from my data frame. I used the "Athlete" and "Country".

```python
best_time_results = df.groupby("Athlete") ["Rank"].min()
best_time_results
```

To use the split-apply-combine methodology to analyze my dataframe, I used the df.groupby() syntax to analyze the rank of the athletes in the dataframe.
